```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
retail_shopping_basket = {
    'ID': [1, 2, 3, 4, 5, 6],
    'Basket': [
        ['Beer', 'Diaper', 'Pretzels', 'Chips', 'Aspirin'],
        ['Diaper', 'Beer', 'Chips', 'Lotion', 'Juice', 'BabyFood', 'Milk'],
        ['Soda', 'Chips', 'Milk'],
        ['Soup', 'Beer', 'Diaper', 'Milk', 'IceCream'],
        ['Soda', 'Coffee', 'Milk', 'Bread'],
        ['Beer', 'Chips']
    ]
}


retail = pd.DataFrame(retail_shopping_basket)
pd.options.display.max_colwidth = 60
print(retail)
```

```python
retail_id = retail.drop('Basket', axis=1)
retail_basket = retail.Basket.str.join(',')
retail_basket
```

```python
retail_basket = retail_basket.str.get_dummies(',')
retail = retail_id.join(retail_basket)     
retail
```

```python
retail_bool = retail.drop('ID', axis=1).astype(bool)
frequent_itemsets_2 = apriori(retail_bool, use_colnames=True)
frequent_itemsets_2
```

只考虑支持度(X>Y),[Beer,Chips]和[Beer,Diper]都很频繁，哪一种更相关？

```python
association_rules(frequent_itemsets_2,metric='lift')
```

结论：[Diaper，Beer]更相关
